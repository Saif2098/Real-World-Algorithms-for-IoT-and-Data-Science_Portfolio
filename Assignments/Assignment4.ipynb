````python
def music_implemented():
    return False

def music_aoa(mic_data_folder, FS, MIC_OFFSETS):
    """MUSIC AoA algorithm. Write your code here.

    Args:
        mic_data_folder: name of folder (without a trailing slash) containing 
                         the mic datafiles `0.csv` through `7.csv` and `config.csv`.
        FS: microphone sampling frequency - 16kHz.
        MIC_OFFSETS: a list of tuples of each microphone's location relative to the center of its mic array. 
                     This list is calculated based on the diameter(0.09218m) and geometry of the microphone array.
                     For example, MIC_OFFSETS[1] is [0.09218*0.5, 0]. If the location of microphone array #i is
                     [x_i, y_i]. Then [x_i, y_i] + MIC_OFFSETS[j] yields the absolute location of mic#j of array#i.
                     This is provided for your convenience and you may choose to ignore.

    Returns:
        A list of incoming signal's polar angles for all mic arrays.

    """  
    aoas = [0, 0, 0, 0, 0, 0, 0, 0] # Please comment out this line after you implement the function
    # Your return value should be the incoming signal's polar angle (between 0 and 2pi)
    return aoas

def delay_and_sum_aoa(mic_data_folder, FS, MIC_OFFSETS):
    """Delay and sum AoA algorithm. Write your code here.

    Args and output same as previous.
    
    """
    # ######################
    def delayed_correlation_with_interpolation(signal1, signal2, shift, fs=16000):
      """
      Calculates the correlation between two sound signals, where the signals are first delayed
      by a non-integer time tau.
      
      Arguments:
      signal1 -- numpy array containing the first sound signal
      signal2 -- numpy array containing the second sound signal
      tau -- sample delay in seconds
      fs -- sampling rate of the signals in Hz (default is 16 kHz)
      
      Returns:
      Correlation between the delayed signals
      """
    
      # Create an interpolator for the signal
      interpolator = interp1d(np.arange(len(signal2)), signal2, kind='linear', fill_value="extrapolate")
      
      # Create an array of new indices for the shifted signal
      new_indices = np.arange(len(signal2)) + shift

      # Use the interpolator to interpolate the signal at the new indices
      shifted_signal = interpolator(new_indices)
      
      # Calculate the correlation between the delayed signals
      corr = np.correlate(signal1, shifted_signal)
      return corr

    ########################

    def sample_delay_vector(theta, fs=16000, num_sensors=6):
      '''
        All time delays are claculated wrt
        the center of the hexagon
      '''
      D = 0.09218/2.0
      v = 343.0              # speed of sound in air in m/s
      CONST = -D * fs / v
      time_delay = CONST * np.array([np.cos(n*(np.pi/3)-theta) for n in range(num_sensors)])

      # reorder

      time_delay = [time_delay[1], time_delay[0], time_delay[5], time_delay[4], time_delay[3], time_delay[2]]

      reference_delay = time_delay[0]

      time_delay = np.array([ele - reference_delay for ele in time_delay])

      return time_delay
    ########################

  

    aoas = []
    
    path = os.path.join(os.getcwd(), mic_data_folder)

    for i in range(0,8):
      # read file
      filename = os.path.join(path, f"{i}.csv")
      df = pd.read_csv(filename, header=None)
      # create angles with 2 degree space
      angles = np.arange(0, 2*np.pi, 0.8*np.pi)
      # mic_0 signal
      mic_0 = df.iloc[:, 0].values
      # mic_0 = signal.resample(mic_0, len(mic_0)*15)

      corr_array = []
      
      # Find correlation of mic_0 with all other mics in this loop
      for i in range(len(angles)):
        # sum correlatin values in this variable 
        corr = 0
        # calculate the time delay vector for angles[i]
        time_delay = sample_delay_vector(angles[i])
        
        # find correlation between mic_0 and mic_j by also taking into account the non-linear delay
        for j in range(1, len(time_delay)):
          # mic_j signal
          mic_j = df.iloc[:, j].values
          # mic_j = signal.resample(mic_j, len(mic_j)*15)
          # corr += delayed_correlation(mic_0, mic_j, time_delay[j])
          corr += delayed_correlation_with_interpolation(mic_0, mic_j, time_delay[j])

        corr_array.append(corr)

      aoas.append(angles[np.argmax(corr_array)])

      # print(np.array(aoas)*180/np.pi)

    aoas = [n+2*np.pi if n<0 else n for n in aoas]

    # Your return value should be the incoming signal's polar angle (between 0 and 2pi)
    return aoas 


def aoa_localization(aoas, mic_data_folder):
    """AoA localization algorithm. Write your code here.

    Args:
        aoas: output for aoa algorithm.
        mic_data_folder: same as previous.

    Returns:
        The user's location in this format (in metres): (L_x, L_y)

    """
    ################
    def delayed_correlation_with_interpolation(signal1, signal2, shift, fs=16000):
      """
      Calculates the correlation between two sound signals, where the signals are first delayed
      by a non-integer time tau.
      
      Arguments:
      signal1 -- numpy array containing the first sound signal
      signal2 -- numpy array containing the second sound signal
      tau -- sample delay in seconds
      fs -- sampling rate of the signals in Hz (default is 16 kHz)
      
      Returns:
      Correlation between the delayed signals
      """
    
      # Create an interpolator for the signal
      interpolator = interp1d(np.arange(len(signal2)), signal2, kind='linear', fill_value="extrapolate")
      
      # Create an array of new indices for the shifted signal
      new_indices = np.arange(len(signal2)) + shift

      # Use the interpolator to interpolate the signal at the new indices
      shifted_signal = interpolator(new_indices)
      
      # Calculate the correlation between the delayed signals
      corr = np.correlate(signal1, shifted_signal)
      return corr
    ###############

    def sample_delay_vector(theta, fs=16000, num_sensors=6):
      '''
        All time delays are claculated wrt
        the center of the hexagon
      '''
      D = 0.09218/2.0
      v = 343.0              # speed of sound in air in m/s
      CONST = -D * fs / v
      time_delay = CONST * np.array([np.cos(n*(np.pi/3)-theta) for n in range(num_sensors)])

      # reorder

      time_delay = [time_delay[1], time_delay[0], time_delay[5], time_delay[4], time_delay[3], time_delay[2]]

      reference_delay = time_delay[0]

      time_delay = np.array([ele - reference_delay for ele in time_delay])

      return time_delay
    ########################

    # not using angles

    path = os.path.join(os.getcwd(), mic_data_folder)
    filename = os.path.join(path, "config.csv")

    df = pd.read_csv(filename, header=None)

    # load microphone data
    mic_data = [pd.read_csv(os.path.join(path, f"{sensor_num}.csv"), header=None) for sensor_num in range(0,8)]
    
    num_rows = df.shape[0]
    
    x_coordinates = df.iloc[:, 0].values
    y_coordinates = df.iloc[:, 1].values

    
    #  range of possible locations
    step = 0.35
    x_range = np.arange(np.min(x_coordinates), np.max(x_coordinates) + np.max(x_coordinates)/2.0, step)
    y_range = np.arange(np.min(y_coordinates), np.max(y_coordinates) + np.max(y_coordinates)/2.0, step)

    corr_max = 0
    corr_max_coordinates = (0,0)

    #  find score of each location pair
    # this x and y serves as the source coordinates
    # print("x range is: {}".format(x_range))
    for x in x_range:
      # print("x is: {}".format(x))
      for y in y_range:
        # find angles for each 

        current_corr = 0

        # this is the microphone x and y coordinates
        for row in range(num_rows):
          x_m = int(df.iloc[row, 0])
          y_m = int(df.iloc[row, 1])

          # find difference
          x_diff = x - x_m
          y_diff = y - y_m

          angle = 0

          # account for nearfield conditions
          if np.sqrt(x_diff**2 + y_diff**2) < 0.2:
            continue
          if x_diff == 0 or y_diff==0:
            continue

          if x_diff > 0 and y_diff > 0:
            angle = np.tan(y_diff/x_diff)
          elif x_diff < 0 and y_diff > 0:
            angle = np.pi - np.tan(-y_diff/x_diff)
          elif x_diff > 0 and y_diff < 0:
            angle = 2*np.pi - np.tan(-y_diff/x_diff)
          else:
            angle = np.pi + np.tan(y_diff/x_diff)

          # find the time delay
          time_delay_array = sample_delay_vector(angle)
          
          # find the correlation
          mic_0 = mic_data[row].iloc[:, 0].values
          for j in range(1, len(time_delay_array)):
            # mic_j signal
            mic_j = mic_data[row].iloc[:, j].values
            # current_corr += delayed_correlation(mic_0, mic_j, time_delay_array[j])
            current_corr += delayed_correlation_with_interpolation(mic_0, mic_j, time_delay_array[j])

        if current_corr > corr_max:
          corr_max = current_corr
          corr_max_coordinates = (x,y)





    # Your return value should be the user's location in this format (in metres): (L_x, L_y)
    return (corr_max_coordinates) 
````
